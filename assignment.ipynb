```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_func(a,x,b):
  return a*x + b
```

```python
def non_linear_func(x):
  return x**2
```

```python
def exp_func(x):
  return 1/(1+np.exp(-x))
```

```python
def exp_func02(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
def exp_func03(a,x,b):
  u = a*x + b
  return  1/(1+np.exp(-u))
```

```python
def exp_func04(a,x,b):
  u = a*x + b
  return  (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))
```

```python
  def exp_func05(a,x,b):
   u = a*x + b
   return 1/(1+np.exp(-u))
```

```python
def exp_func06(a1,a2,y1,y2,b):
   u = a1*y1 + a2*y2 + b
   return 1/(1+np.exp(-u))
```

```python
a = 5
b = 1
x = np.arange(-10,10,0.1)
```

1st function: y = wx+b where w,b constant

```python
y = linear_func(a,x,b)

plt.plot(x,y, color='green' ,label= 'y=5x+1')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-20, 30)
plt.title(f"linear function for y = 5x+1")
plt.legend()
plt.grid(True)
plt.show()
```

2nd function: Y =  x^2

```python
y = non_linear_func(x)

plt.plot(x,y, color='green',label= 'y=x^2')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-20, 30)
plt.title(f"non linear function for y = x^2")
plt.legend()
plt.grid(True)
plt.show()
```

3rd Function:  Y = 1/(1+e^(-x))

```python
y = exp_func(x)

plt.plot(x,y, color='green' ,label= '1/(1+e^(-x))')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(0, 1)
plt.title(f"exponential function for Y = 1/(1+e^(-x))")
plt.legend()
plt.grid(True)
plt.show()
```

4th Function: Y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
y = exp_func02(x)

plt.plot(x,y, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"exponential function for Y = (e^x - e^(-x)) / (e^x + e^(-x))")

plt.show()
```

5th Function: Y = 1/(1+e^(-u)) where u = wx+b


```python
y = exp_func03(a,x,b)

plt.plot(x,y, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"exponential function for  Y = 1/(1+e^(-u))")

plt.show()
```

6th function: Y = (e^u - e^(-u)) / (e^u + e^(-u)) where u = wx+b


```python
y = exp_func04(a,x,b)

plt.plot(x,y, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"exponential function for  Y = (e^u - e^(-u)) / (e^u + e^(-u))")

plt.show()
```
